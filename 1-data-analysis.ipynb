# Reading data

```python
import numpy as np
import pandas as pd

train_data = pd.read_csv("/kaggle/input/california-homelessness-prediction-challenge/train.csv")

train_data.head()
```

# Exploring data
We need to know what kind of information we're working with! Let's explore the columns.

```python
# Column names
print(", \n".join(train_data.columns))
```

```python
# Total count of null values per column
train_data.isnull().sum()
```

Alright! So we don't have any _missing_ values.  
Do we have any zeroes? And are they valid values for the column?

```python
# Total count of zeros per column
(train_data == 0).sum()
```

We are seeing zeros for the following columns:
- HOMELESS_RATE
- RACE_BLACK_NH_PCT
- RACE_NATIVE_NH_PCT
- RACE_PACIFIC_NH_PCT
- MULTI_PERSON_NONFAMILY_HH_PCT  
Let's see what these values represent and make sure they aren't invalid / noise.

```python
train_data[["HOMELESS_RATE", "RACE_BLACK_NH_PCT", "RACE_NATIVE_NH_PCT", "RACE_PACIFIC_NH_PCT", "MULTI_PERSON_NONFAMILY_HH_PCT"]].head()
```

Based on the Kaggle data description, each column is census data from particular regions, meaning this is demographic data. It would make sense that some places simply have no population of certain race/ethnicity groups, or perhaps the census didn't cover those groups. In any case, it seems that all of these columns are valid at zero, so we don't need to remove them.

## Visualizing

Let's plot each column on small bin graphs to visualize the distribution and possible outliers.

```python
import matplotlib.pyplot as plt
import math

graph_count = len(train_data.columns)
col_count = 4
row_count = math.ceil(graph_count / col_count)

fig, axes = plt.subplots(row_count, col_count)
fig.set_size_inches(col_count*3, row_count*3)
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.25, hspace=0.4)
axes = axes.ravel()
for i in range(graph_count):
    axes[i].hist(train_data[train_data.columns[i]], bins=15)
    axes[i].set_title(train_data.columns[i])
```

Wonderful! There's no glaring outliers and the data seems to follow normal-ish distributions. This data seems very clean.
